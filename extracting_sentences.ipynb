```python
import pandas as pd
```

```python
df = pd.read_json('/content/drive/MyDrive/messages.json')
```

```python
df.head(40)
```

```python
# df.drop(['userId','exportDate'], axis=1)
convo = df['conversations']
```

```python
ans= pd.DataFrame()

convo_iter = convo[2]
student = []
teacher = []
for i in range(len(convo_iter['MessageList'])):
    if convo_iter['MessageList'][i]['from'] == '8:live:.cid.4933a22bf564744' and '<' not in convo_iter['MessageList'][i]['content']:
      teacher.append(convo_iter['MessageList'][i]['content'])
      if convo_iter['MessageList'][i-1]['from'] != '8:live:.cid.4933a22bf564744' and '<' not in convo_iter['MessageList'][i-1]['content']:
        student.append(convo_iter['MessageList'][i-1]['content'])
ans.append( pd.DataFrame(
    {'student': student,
    'teacher': teacher[0:len(student)]
    }), ignore_index=True)
```
